```python
"""
Below is the document Report link for Analysis document :
https://docs.google.com/document/d/1or6AdBaRysWuH1Gz5MPhD129vMvGdktl/edit?usp=sharing&ouid=107857076080323012434&rtpof=true&sd=true
"""


from bs4 import BeautifulSoup
from selenium  import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
import time

Names = []
Reviews = []
Dates = []
Cities = []

url = """https://www.flipkart.com/apple-iphone-15-blue-128-gb/product-reviews/itmbf14ef54f645d?pid=MOBGTAGPAQNVFZZY&lid=LSTMOBGTAGPAQNVFZZYQRLPCQ&marketplace=FLIPKART"""
driver = webdriver.Chrome()
driver.get(url)

for i in range(2,5):

    time.sleep(2)
    soup = BeautifulSoup(driver.page_source,"html.parser")

    Names = soup.find_all("p",{"class":"_2NsDsF AwS1CA"})
    for i in Names :
        names.append(i.text)

    Reviews = soup.find_all("div",{"class":"ZmyHeo"})
    for i in Reviews :
        reviews.append(i.text)

    Cities = soup.find_all("p",{"class":"MztJPv"})
    for i in Cities:
        cities.append(i.text)

    Dates = soup.find_all("p",{"class":"_2NsDsF"})
    for i in Dates:
        dates.append(i.text)

    for i in dates :
        act_dates = dates[1::2]

    time.sleep(2)

    driver.find_element(By.XPATH,"""/html/body/div[1]/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]""").click()
        

```

```python
import pandas as pd
df = pd.DataFrame({"Names":names, "Cities":cities,"Reviews":reviews,"Dates":act_dates})
df
```

```python
df["Reviews"] = df["Reviews"].str.lower()
df
```

```python
#nltk and TextBlob libraries for Tokenizing = Break down into list of sentences into different sentences
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize   #Similarly, word tokenize breaks down diff words.
from textblob import TextBlob
import string

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
#To tokenize the words in the Reviews
df['Reviews'] = df['Reviews'].fillna("")
df["Tokenized_Reviews"] = df["Reviews"].apply(sent_tokenize)
df
```

```python
"""
To find the polarity of each of sentence in Reviews column, we use statistics.
Then, we find the mean of polarity of each Review(if there are more than one polarity/more than one sentence after sent_tokenization)
Lastly we find the Avergae of Complete Polarity.
"""

from statistics import mean

def find_polarity(sentences):
    return [TextBlob(sentence).sentiment.polarity for sentence in sentences]

df['Polarity'] = df['Tokenized_Reviews'].apply(find_polarity)

def calculate_avg_polarity(polarities):
    return mean(polarities) if polarities else 0

df['Average_Polarity'] = df['Polarity'].apply(calculate_avg_polarity)

df.head(10)
```
